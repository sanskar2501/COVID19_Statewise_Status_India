### Corona Virus Live Tracking

```python
import pandas as pd
```

```python
## read data from https://www.mohfw.gov.in/
df = pd.read_json("https://www.mohfw.gov.in/data/datanew.json")
df
```

```python
# we don't need last row and sno column
df = df.iloc[:-1,1:]
```

```python
df.tail()
```

```python
# importing libraries for visualization
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
total_states = np.arange(len(df))
total_states
```

### 1. Total Positive Cases in India based in States

```python
plt.figure(figsize=(15,10))
plt.barh(total_states, df["positive"])
plt.yticks(total_states, df['state_name'])
plt.xlabel("Number of Positive Cases")
plt.title("Number of Positive cases based on States")
```

### 2. Total Active Cases in India based in States

```python
plt.figure(figsize=(15,10))
plt.barh(total_states, df["active"])
plt.yticks(total_states, df['state_name'])
plt.xlabel("Number of Active Cases")
plt.title("Number of Active cases based on States")
```

### 3. Total Cured Cases in India based in States

```python
plt.figure(figsize=(15,10))
plt.barh(total_states, df["cured"])
plt.yticks(total_states, df['state_name'])
plt.xlabel("Number of Cured Cases")
plt.title("Number of Cured cases based on States")
```

### 4. Total Deaths in India based in States

```python
plt.figure(figsize=(15,10))
plt.barh(total_states, df["death"])
plt.yticks(total_states, df['state_name'])
plt.xlabel("Number of Deaths")
plt.title("Number of Deaths based on States")
```

### 5. Active vs Positive vs Cured

```python
df1 = df.iloc[:,1:4]
df1.head()
```

```python
df1.plot.barh(figsize=(15,10))
plt.yticks(total_states, df['state_name'])
plt.xlabel("Number of Cases")
plt.title("Active vs Positive vs Cured")
```
